```python
for index, row in df.iterrows():
    if(row['Stress'] > 0):
        row['symptom'] = 1
    if(row['PTSD'] > 0):
        row['symptom'] = 2
    if(row['Speech'] > 0):
        row['symptom'] = 3
    if(row['Anxiety'] > 0):
        row['symptom'] = 4
    if(row['Depression'] > 0):
        row['symptom'] = 5
    if(row['Headache'] > 0):
        row['symptom'] = 6
    if(row['Sleep'] > 0):
        row['symptom'] = 7
    if(row['Audiology'] > 0):
        row['symptom'] = 8
    if(row['Vision'] > 0):
        row['symptom'] = 9
    if(row['Neurologic'] > 0):
        row['symptom'] = 10
    if(row['Alzheimer'] > 0):
        row['symptom'] = 11
    if(row['Cognitive'] > 0):
        row['symptom'] = 12
    if(row['PCS'] > 0):
        row['symptom'] = 13
    if(row['Endocrine'] > 0):
        row['symptom'] = 14
    
```
